```python
import pandas as pd
import numpy as np
```

```python
frame = pd.DataFrame({'id':['A','B','C','D','E'],'price':[11,12,13,14,15]})
```

```python
frame
```

```python
frame2 = pd.DataFrame( {'id': ['A','B','C','D'],
                        'color':['white','red','red','black']})
frame2
```

```python
pd.merge(frame,frame2)
```

We can also give arg -> on = 'Column name' in merge() , it can be either a list or individual features

```python
frame1 = pd.DataFrame( {'id': ['ball','pencil','pen',',mug','ashtray'],
                        'color': ['white','red','red','black','green'],
                        'brand': ['OMG','ABC','ABC','POD','PPOD']})
frame2 = pd.DataFrame( {'id':['pencil','pencil','ball','pen'],
                        'brand': ['OMG','POD','ABC','POD']})
frame2
frame1
```

```python
pd.merge(frame1,frame2,on='id')
```

```python
pd.merge(frame1,frame2,left_on='id',right_on='brand')
```
